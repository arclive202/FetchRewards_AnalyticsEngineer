## Data Transformation & Cleaning

```python
#Function to treat values which are within a dictionary

from datetime import datetime


def treatment(cell):
  if isinstance(cell, dict):
      if '$oid' in cell:  
          return cell['$oid']
      elif '$date' in cell:
          # Convert the timestamp (milliseconds) to a datetime string
          return datetime.utcfromtimestamp(cell['$date'] / 1000).strftime('%Y-%m-%d %H:%M:%S')
  return cell

def series_treatment(row):
    if isinstance(row, dict):
        return pd.Series({'cpg_id': row.get('$id', None), 'ref': row.get('$ref', None)})
    else:
        return pd.Series({'id': None, 'ref': None})

def extract_data(df):
  return df.applymap(treatment)



def null_values_treatment(df):
    # Fill NaN values for string columns with 'unknown'
    string_columns = df.select_dtypes(include=['string']).columns
    df[string_columns] = df[string_columns].fillna('unknown')

    # Fill NaN values for numeric columns (float and int) with 0
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
    df[numeric_columns] = df[numeric_columns].fillna(0)

    # Return the updated DataFrame
    return df
```

```python
#extract the values of cpg into 2 separate columns in the brand dataframe
cpg_df = pd.DataFrame(brands_df['cpg'])
brands_df[['cpg_id', 'ref']] = cpg_df['cpg'].apply(series_treatment)

#Extract all the data with $oid, $date etc.
users = extract_data(users_df)
brands = extract_data(brands_df)
receipts = extract_data(receipts_df)
```

```python
# Renaming the _id columns to clearly differentiate between IDs in the dataframes

users = users.rename(columns={'_id': 'userId'})
brands = brands.rename(columns={'_id': 'brandId','name': 'brandName'})
receipts = receipts.rename(columns={'_id': 'receiptId'})
```

### Creating ReceiptsItemList

<br/> Extracting the 'receiptItemslist' from the receipts dataframe in order to create a new table with its own unique UUID and holding the data about the item lists along with the receiptID as the foreign key

```python
#creating a new receiptsItemList dataframe that will unpack the data from rewardsReceiptItemList
import uuid

def createReceiptItemList(df):
  unpacked_data = []
  for index, row in df.iterrows():
    receiptID = row['receiptId']
    items = row['rewardsReceiptItemList']
    if isinstance(items, list):
      for item in items:
        partnerItemId = item['partnerItemId']
        new_uuid = str(uuid.uuid5(uuid.NAMESPACE_DNS,f"{receiptID}-{partnerItemId}"))
        unpacked_data.append({'receiptsItemListId':new_uuid,'receiptId': receiptID, **item})
    # else:
    #   print(f'Warning: Row {index} does not have valid Items List \n {items} \n')
  return pd.DataFrame(unpacked_data)

receiptsItemList_df = createReceiptItemList(receipts)

len(receiptsItemList_df)
```

### Duplicate Checks

These are basic duplicates checks performed on Primary keys of the table and other custom columns based on business needs to ensure data looks good. <br/>

```python
def check_duplicates(df, primary_key=None, duplicate_columns=None):
    results = []
    
    # Check duplicates on primary key
    if primary_key:
        pk_duplicates = df.duplicated(subset=[primary_key]).sum()
        results.append({'Check': 'Primary Key Duplicates', 'Column(s)': primary_key, 'Count': pk_duplicates})
    
    # Check duplicates on specific columns
    if duplicate_columns:
        col_duplicates = df.duplicated(subset=duplicate_columns).sum()
        results.append({'Check': 'Duplicate Columns', 'Column(s)': ', '.join(duplicate_columns), 'Count': col_duplicates})
    
    return pd.DataFrame(results)
```

### Non Null Checks

These are custom checks to take a look at the dataframe to see what values of the dataframe have null values in them and treat them in the following section accordingly in order to prevent errors during data transformations further

```python
def check_non_nulls(df):
    null_counts = df.isnull().sum()
    null_percentages = (df.isnull().mean() * 100).round(2)
    
    report = pd.DataFrame({
        'Column': null_counts.index,
        'Null Count': null_counts.values,
        'Null Percentage': null_percentages.values
    })
    
    return report
```

### Data Recency checks

These checks are usually performed on fields like createdDate and modifiedDate to keep a recency on the data because on data model and pipelines one of the biggest issues could arise out of stale data leading to wrong reports

```python
def data_recency_check(df, date_columns):
    threshold_date = pd.to_datetime(datetime.now().date() - timedelta(days=1))
    results = []
    for column in date_columns:
        col_val = df[df[column].notnull()][[column]]
        max_date = pd.to_datetime(col_val[column].max())
        is_valid = max_date >= threshold_date if pd.notnull(max_date) else False
        results.append({
                    'Column': column,
                    'Max Date': max_date,
                    'Threshold Date': threshold_date,
                    'Valid': is_valid
                })
    report = pd.DataFrame(results)
    
    return report
```
